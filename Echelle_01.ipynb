## SIMBAD

Input: coords_raw.txt (coordinates from the concordance file)<br>
SIMBAD coordinate query - 10 arc sec (with http://simbad.u-strasbg.fr/simbad/sim-fout)<br>
Output: coords_simbad.txt<br>
<br>
Major problems<br>
(1) No object found from SIMBAD<br>
(2) Multiple objects found (6,072 out of 33,909) - use the smallest angulation for now (to begin with)<br>
(3) The same object is found for different Echelle objects (e.g., HIP013117A, HIP013117AB, HIP013117B)

```python
in_file = open("coords_simbad.txt", "r")
out_file_1 = open("coords_simbad_output_1.txt", "w")
out_file_2 = open("coords_simbad_output_2.txt", "w")

i = 1
for line in in_file:
    if re.match("(.*)No astronomical object found(.*)", line):
        print >> out_file_1, i, "|", "~|~|~|~ ~ ~ ~ ~ ~|~ ~|~ ~|~|~|~|~|~|~\n",
#        print >> out_file, i, "|", "|||||||||||\n",
#        print >> out_file, i, "|", line,
        i = i + 1
    elif re.match("1(\|| \|)", line):
        print >> out_file_1, i, "|", line,
        i = i + 1
    elif re.match("[2-9](\|| \|)", line):
        print >> out_file_2, i, "|", line,

in_file.close()
out_file_1.close()
out_file_2.close()

# out_file_1 - rank 1 objects or no object found
# out_file_2 - rank 2-9 objects
```

 <br>
 For SIMBAD objects found and chosen to be used, polish and store the data.

```python
names = ['target_id','rank','sep','simbad_identifier','coords1','coords2','pm','plx','rv','z','magB','magV','sp_type']
```

```python
df1 = pd.read_csv('coords_simbad_output_1.txt', sep='|', header=None, names=names, skipinitialspace=True)
len(df1)
```

```python
df1.head(8)
```

```python
df1['rah'],df1['ram'],df1['ras'],df1['decd'],df1['decm'],df1['decs'] = df1['coords1'].str.split(' ',5).str
df1['ra_d'],df1['dec_d'] = df1['coords2'].str.split(' ',1).str
df1['pm_ra'],df1['pm_dec'] = df1['pm'].str.split(' ',1).str
```

```python
df1.head(8)
```

```python
df1['ra'] = df1['rah'].astype(str).str.cat(df1['ram'].astype(str), sep=':')
df1['ra'] = df1['ra'].astype(str).str.cat(df1['ras'].astype(str), sep=':')

df1['dec'] = df1['decd'].astype(str).str.cat(df1['decm'].astype(str), sep=':')
df1['dec'] = df1['dec'].astype(str).str.cat(df1['decs'].astype(str), sep=':')
```

```python
df1.head(8)
```

```python
df1.drop(['coords1','coords2','pm','rah','ram','ras','decd','decm','decs'], axis = 1).head(8)
```

```python
df1 = df1[['target_id','rank','sep','simbad_identifier','ra','pm_ra','dec','pm_dec','ra_d','dec_d','plx','rv','z','magB','magV','sp_type']]
```

```python
df1.head(8)
```

```python
# remove trailing whitespaces
df1.loc[:,('simbad_identifier')] = df1.loc[:,('simbad_identifier')].str.strip()
df1.loc[:,('pm_ra')] = df1.loc[:,('pm_ra')].str.strip()
df1.loc[:,('dec')] = df1.loc[:,('dec')].str.strip()
df1.loc[:,('pm_dec')] = df1.loc[:,('pm_dec')].str.strip()
df1.loc[:,('dec_d')] = df1.loc[:,('dec_d')].str.strip()
df1.loc[:,('rv')] = df1.loc[:,('rv')].str.strip()
df1.loc[:,('z')] = df1.loc[:,('z')].str.strip()
df1.loc[:,('magB')] = df1.loc[:,('magB')].str.strip()
df1.loc[:,('magV')] = df1.loc[:,('magV')].str.strip()
df1.loc[:,('sp_type')] = df1.loc[:,('sp_type')].str.strip()
```

```python
df1.head(8)
```

```python
# df_simbad_selected.csv - SIMBAD objects found and chosen to be used
df1.to_csv('df_simbad_selected.csv', index=False)
```
